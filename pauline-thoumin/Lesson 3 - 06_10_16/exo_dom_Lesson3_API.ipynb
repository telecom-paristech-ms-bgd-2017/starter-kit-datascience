```python
############################ Devoir sur les API pour le 14/10/16 ##################################

import requests
import bs4
import os
from pandas import *

url = 'https://gist.github.com/paulmillr/2657075'

result = requests.get(url)
soup = bs4.BeautifulSoup(result.content, 'html.parser')
#a = soup.find_all("table") # nom de balises
a = soup.find_all("tr")
# soup.find_all("a") # dictionnaire : clé href

rang = [] #a[1].find('th').text.replace('</th>','').replace('<th scope="row">','') 
user = [] #a[1].find('td').text.split('>')[0].split(' ')[0]
contribs = [] #int(a[1].find_all('td')[1].text.replace('<td>','').replace('</td>',''))
location = [] #a[1].find_all('td')[2].text.replace('<td>','').replace('</td>','')
url = [] #a[1].find('a')['href']
    
for i in range(1, len(a)) :
    rang.append(a[i].find('th').text.replace('</th>','').replace('<th scope="row">',''))
    user.append(a[i].find('td').text.split('>')[0].split(' ')[0])
    contribs.append(int(a[i].find_all('td')[1].text.replace('<td>','').replace('</td>','')))
    location.append(a[i].find_all('td')[2].text.replace('<td>','').replace('</td>',''))
    url.append(a[i].find('a')['href'])



tab=pandas.concat([DataFrame(rang), DataFrame(user), DataFrame(contribs), DataFrame(location),DataFrame(url)], axis=1)
tab.columns = ["Rang","user", "contribs", "location", "url"]
print tab.head


```
